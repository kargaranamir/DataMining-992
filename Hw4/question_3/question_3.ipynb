## Part 1
- load data 
- determine threshold and ratio of missing data
- drop the features with missing data ratio over threshold
- show missing data ratio of columns (before/after)

```python
df = pd.read_csv('dimensionality_reduction_cbb.csv')

threshold = 0.8
norm_null = df.isnull().sum()/len(df)
dropped_features = norm_null.loc[norm_null > threshold].index
df = df.drop(dropped_features,axis=1)
norm_null2 = df.isnull().sum()/len(df)

print(f"**old missing data ratio of columns**:\n{norm_null},\n**new missing data ratio of columns**:\n{norm_null2}") 
```

## Part 2
- load data, make a copy
- find non numeric columns an convert them to numeric
- compute variance
- drop columns with less variance than threshold
- back to the original form 
- report

```python
df = pd.read_csv('dimensionality_reduction_cbb.csv')
df_copy = df.copy()

for i in list(df.columns):
    df[i].fillna(df[i].mode()[0],inplace=True)

df_none_num = df.select_dtypes(include=np.object_)

# non_numeric -> categorical -> numeric
for i in list(df_none_num.columns):
    df[i] = df[i].astype('category')
    df[i] = df[[i]].apply(lambda x: x.cat.codes)
    df[i] = df[i].astype('int64')
    

threshold = 10
variance = df.var()
dropped_features = norm_null.loc[variance < threshold].index
df = df.drop(dropped_features,axis=1)
variance2 = df.var()

# back to the original 
for i in list(df_none_num.columns):
    if i in df.columns:
        df[i] = df_copy[i]

print(f"**old variance of columns**:\n{variance},\n**new variance of columns**:\n{variance2}")
```

## Part 3
- load data
- set threshold and find | high correlated columns |
- report list of pairs
- question doesnt want to delete any columns! and just says report them with their correlation.

```python
df = pd.read_csv('dimensionality_reduction_cbb.csv')

list_of_pairs = []
threshold = 0.7
correlation = df.corr().abs()
correlated_features = np.where(correlation > threshold)
correlated_features = list(zip(*correlated_features))
for i,j in correlated_features:
    if i!=j and i<j:
        x = correlation.columns[i]
        y = correlation.columns[j]
        list_of_pairs.append((x, y, correlation.loc[correlation.index == x][y][0]))
list_of_pairs
```
