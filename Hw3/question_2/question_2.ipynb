## part 1:
- `day` represented as diffrenet coulmns and `year` and `month` represented as seprated columns -> we should merge them as `date`, especially the `day`! at least `day` should be a single coulmn not 31s.
- extra coulmns like `X`
- managing Non values
-  numeric data should have a numeric type and inconsistency should be reolved in `PrecipitationIn` column (`T` is equal to `0`)
- `date` is a good term to be as an  index
- percentage is between `0` to `100`
- This unequlaity should be correct in all related columns : `Max >= Mean >= min`

## load dataset

```python
df = pd.read_csv('weather.csv', index_col='Unnamed: 0')
df
```

## rename X columns to integers

```python
dict_X = {'X' + str(i):i for i in range(1,32)}
df.rename(columns=dict_X, inplace=True)
```

## convert date
- Melt data 
- convert (year, month, day) -> datetime form (date)
- drop extra columns
- drop null values

```python
df = pd.melt(df, id_vars = ['X', 'year', 'month', 'measure'], var_name='day')
df['date'] = df[['year', 'month', 'day']].apply(lambda x: pd.to_datetime(f'{x.year} {x.month} {x.day}', errors='coerce'), axis=1)
df.drop(['X', 'year', 'month', 'day'], inplace = True, axis = 1)
df.dropna(inplace=True)
df
```

## pivot (melt^{-1}) -> getting back to normal

```python
df = df.pivot(index = ['date'], columns = 'measure')
df.columns = [col[1] for col in df.columns]
df
```

## final process
- Trace amounts of precipitation are abbreviated by the capital letter "T", often placed in parenthesis (T). If you must convert a trace to a numerical amount, it would equal 0.00 (https://www.thoughtco.com/what-is-trace-of-precipitation-3444238)

- maximum amount of water vapor in the air will produce a relative humidity (RH) of 100 percent. Max.Humidity has a one row problem.
- Max >= Mean >= min -> Mean.VisibilityMiles has a one row problem

```python
df['PrecipitationIn'] = df['PrecipitationIn'].str.replace('T', '0')

# convert columns to numeric except date and events
cols = [i for i in df.columns if i not in ["date","Events"]]
for col in cols:
    df[col] = pd.to_numeric(df[col])

df.loc[(df['Max.Humidity'] > 100, 'Max.Humidity')] = 100
df.loc[(df['Mean.VisibilityMiles'] < df['Min.VisibilityMiles'], 'Mean.VisibilityMiles')] = float('NaN')
```

## export csv

```python
df.to_csv('cleaned_weather.csv')
```
