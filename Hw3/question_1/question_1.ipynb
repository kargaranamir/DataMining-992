## part 1 (load data show 7 first rows)

```python
df = pd.read_csv('entire_library_have_fun.csv')
df.head(7)
```

## Part 2 (compute mean of popularity based on Explict valuse and get the arg_max)

```python
p_mean = dict()
p_mean['True_Explicit'] = df[df['Explicit'] == True][['Popularity']].mean()[0]
p_mean['False_Explicit'] = df[df['Explicit'] == False][['Popularity']].mean()[0]

max(p_mean, key=p_mean.get)
```

## Part 3 (result of dec ordering of TN over Ceremony band will be computed as Ceremony_Des_TN)

```python
Ceremony_Des_TN = df[df['Artist Name(s)'] == 'Ceremony'].sort_values(by=['Track Number'], ascending = False)
Ceremony_Des_TN
Ceremony_Des_TN[['Artist Name(s)', 'Track Number']] #proof
```

## Part 4 (make a sparse matrix and apply a sum function over genres)

```python
df['Artist Genres'].str.get_dummies(',').astype(bool).sum(axis=0).sort_values(ascending=False)[:10]
```

## Part 5 ( show rows if they have a 'rock' in genres as  a substring)

## Part 8

-  we can calculate different type of correlation metrics, but as we see their correspponding correlation between `Energy` and `Danceability` is little.
- so we try to figure histograms and it seems very intersting! `Energy` and `Danceability` have a direct relationship for those song with `Danceability` less than approx `0.5` and inverse relationship for the others.
- also when the songs have more `Energy` their `Danceability` is high.

```python
for m in ['pearson', 'kendall', 'spearman']:
    print(f" method: {m} correlation: {df.corr(method=m)[['Danceability']].loc['Energy'][0]}")
```

```python
plt.hist(df.Danceability, weights=df.Energy)
plt.xlabel('Danceability')
plt.ylabel('Energy')
```

```python
plt.hist(df.Energy, weights=df.Danceability)
plt.xlabel('Energy')
plt.ylabel('Danceability')
```
